## Ejemplo: integral vectorial de superficie

Sea $V$ la superficie con forma de vaso formada por el cilindro $x^2+y^2=4$ con $1\le z\le 6$ y una tapa plana a altura $z=1$ que "cierra" el cilindro (la cara inferior del vaso). No hay tapa superior (como es usual en los vasos). Orientar esta superficie con vectores normales "saliendo" del vaso.

**Problema.** Calcular la integral vectorial $$\iint_V (x^2,0,z)\cdot\textrm{d}S$$

---

[Gráfica de la superficie](https://www.geogebra.org/m/k2khwn8r)

---

Para parametrizar esta superficie, habrá que separarla en dos trozos: una superficie $V_1$, la cara lateral del cilindro; y otra superficie $V_2$, el disco que hace de cara inferior. Finalmente, se combinará todo con la relación $$\iint_V (x^2,0,z)\cdot\textrm{d}S = \iint_{V_1} (x^2,0,z)\cdot\textrm{d}S + \iint_{V_2 }(x^2,0,z)\cdot\textrm{d}S$$

---

### La cara lateral

En coordenadas cilindricas, la cara lateral del cilindro se puede describir con las relaciones $$\begin{cases}  r=2 \\ 0\le\theta\le 2\pi \\ 1\le z\le 6\end{cases}$$

Por lo tanto una parametrización de $V_1$ está dada por $$\sigma(\theta,z) = (2\cos(\theta), 2\operatorname{sen}(\theta), z)$$ con $0\le\theta\le 2\pi$ &nbsp;y &nbsp;$1\le z\le 6$.

Se tiene

$$\begin{gathered}
\sigma_\theta &=& (-2\operatorname{sen}(\theta),2\cos(\theta),0)\\[3mm]
\sigma_z &=& (0,0,1)\\[3mm]
\sigma_\theta \times\sigma_z &=& (2\cos(\theta),2\operatorname{sen}(\theta),0)
\end{gathered}$$

Al graficar este vector normal, por ejemplo, en el punto $\theta=0$, $z=1$, vemos que apunta hacia afuera del cilindro, y por lo tanto está con la orientación correcta.

Por lo tanto

$$\begin{gathered}
\iint_{V_1}(x^2,0,z)\cdot\textrm{d}S &=& \int_0^{2\pi}\int_1^6 (4\cos^2 (\theta),0,z)\cdot (2\cos(\theta),2\operatorname{sen}(\theta),0)\,\textrm{d}z\,\textrm{d}\theta\\[3mm]
\iint_{V_1}(x^2,0,z)\cdot\textrm{d}S &=& \int_0^{2\pi}\int_1^6 8\cos^3(\theta)\,\textrm{d}z\,\textrm{d}\theta=40\int_0^{2\pi}\cos^3(\theta)\,\textrm{d}\theta = 0
\end{gathered}$$

---

### La cara inferior

En coordenadas cilíndricas, la cara inferior del cilindro se puede describir con las relaciones $$\begin{cases}0\le r\le 2 \\ 0\le\theta\le 2\pi \\ z=1\end{cases}$$

Por lo tanto una parametrización de $V_2$ está dada por  $$\eta(r,\theta) = (r\cos(\theta),r\operatorname{sen}(\theta),1)$$ con $0\le r\le 2$ &nbsp;y&nbsp; $0\le\theta\le 2\pi$.

Se tiene

$$\begin{gathered}
\eta_r &=& (\cos(\theta),\operatorname{sen}(\theta),0)\\[3mm]
\eta_\theta &=& (-r\operatorname{sen}(\theta),r\cos(\theta),0)\\[3mm]
\eta_r\times\eta_\theta &=& (0,0,r)
\end{gathered}$$

Observamos que este vector normal apunta hacia arriba, pero como la cara es inferior queremos que apunte hacia abajo. Por consiguiente el vector normal que se necesita será su opuesto, $(0,0,-r)$.

Por lo tanto

$$\begin{gathered}
\iint_{V_2} (x^2,0,z)\cdot\textrm{d}S &=& \int_0^2\int_0^{2\pi}(r^2\cos^2(\theta),0,1)\cdot(0,0,-r)\,\textrm{d}\theta\,\textrm{d}r\\[3mm]
\iint_{V_2} (x^2,0,z)\cdot\textrm{d}S &=& \int_0^2\int_0^{2\pi}-r\,\textrm{d}\theta\,\textrm{d}r = \int_0^2-2\pi r\,\textrm{d}r=-4\pi
\end{gathered}$$

---

El resultado es $$\iint_V (x^2,0,z)\cdot\textrm{d}S = 0 +(-4\pi) = -4\pi$$
